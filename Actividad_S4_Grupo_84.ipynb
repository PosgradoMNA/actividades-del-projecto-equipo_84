#Parte 4: Normalización de los datos y aplicación del análisis de componentes principales PCA

Podemos observar que el hecho de que las variables estan altamente correlacionadas significa para el modelo información redudante o repetida. 

```python
cols= ['Estc_sep', 'Edad'] #definimos la lista de variables que nos interesa analizar porque tienen mayor correlación

#creemos un objeto que contiene los estadisticos a utilizar

summary= {"Varianza": new7_df[cols].var()[0:3], 
         "Valor Min": new7_df[cols].min()[0:3], 
         "Valor Max": new7_df[cols].max()[0:3]
} #[0:3] es para que pandas solo imprima valores númericos sin el nombre de cada variable 

pd.DataFrame(summary, index=cols).transpose() #ingresamos los valores en DF y aplicamos transpuesta

```

De la tabla anterior observamos que el valor de varianza para la variable Estc_sep es 5,42e+9, mientras que para edad  es 84,9. 

Podriamos decir que la variable que contiene más información es la de Estc_sep, sin embargo, debemos observar el rango de cada variable.

Poor ejemplo, la variable edad tiene un rango de 21 a 79 y  Estc_sep tiene un rango de -1,65e+5 hasta 9,64e+05.

Esto quiere decir que el valor de varianza se ve afectado por la magnitud de las variables.

Lo anterior puede afectarnos en el análisis de PCA porque es un método que usa las varianzas para decir cual variable es más importante o aporta más información.

Podemos observar todo lo anterior, si buscamos el % de varianza total de las varianzas que representan cada una:

```python
t = new7_df.var().sum()
```

```python
varEdad = new7_df.Edad.var()
varEstc_sep = new7_df.Estc_sep.var()


```

```python
print('Total Varience \t\t\t:', t.round(2) )
print('Varience prop. of Edad\t:', ((varEdad/t)*100).round(2),'%' )
print('Varience prop. of Importe del extracto de cuenta en septiembre\t:', ((varEstc_sep/t)*100).round(2),'%' )
```

```python
#correlación entre las variables
new7_df[cols].corr().round(2)
```

Podemos observar que la variable Importe del extracto de cuenta en septiembre "Estc_sep" representa el 12,47% del total de la varianza. Como dijimos anteriormente, esto no necesariamente significa que el monto sea una variable que aporte más que las otras, porque este resultado se debe a que tiene un rango de valores mucho más alto.

Para poder resolver este problema, debemos escalar los datos. Esto nos va a permitir que todos los datos esten en una misma escala y así evitar que sea vea afectado el resultado del análisis de componentes principales(PCA).



##Normalización 

Usaremos el metodo StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() #crea una variable que almacena el metodo

scaled = scaler.fit_transform(new7_df) #scaled va a almaceenar los valores despues de la estandarización
scaled[:5]
```

```python
#Ingresamos los valores a un DataFrame

scaled_df = pd.DataFrame(scaled, columns= new7_df.columns)

scaled_df.head()
```

```python
summary={"Varianza": scaled_df[cols].var()[0:3], 
         "Valor Min": scaled_df[cols].min()[0:3], 
         "Valor Max": scaled_df[cols].max()[0:3]
} #[0:3] es para que pandas solo imprima valores númericos sin el nombre de cada variable 

pd.DataFrame(summary, index=cols).transpose() #ingresamos los valores en DF y aplicamos transpuesta
```

Podemos observar como todas las variables tienen la misma escala, podemos confirmar esto buscando nuevamente las varianzas y el rango de las variables que estamos analizando.

Una vez escalados los datos, aplicaremos PCA

##PCA

```python
from sklearn.decomposition import PCA
```

```python
pcs = PCA()

pcs_t = pcs.fit_transform(scaled_df) #desde aquí comienza las componentes

pcs_t[:5]
```

Debemos calcular el % de la varianza explicada por cada comonente y el acumulado de la varianza explicada por cada componente

```python
pcsSummary_df = pd.DataFrame({'% varianza explicada':np.round(pcs.explained_variance_ratio_,4)*100,
                              
                              '% varianza acumulada': np.cumsum(pcs.explained_variance_ratio_)*100})

pcsSummary_df
```

Agregaremos etiquetas a cada fila para poder hacer la interpretación más fácil

```python
#crear una lista con los nombres de cada componente 
# labels=[]

# for i in range(len(scaled_df.columns)):
#     labels.append(f'PC{i+1}')
# pcsSummary_df.index=labels


pcs_labels = [f'PC{i + 1}' for i in range(len(scaled_df.columns))]

pcsSummary_df.index = pcs_labels

pcsSummary_df


```

Podemos observar las diferentes componentes principales que son basicamente la descomposición lineales de nuestras variables originales.

Estas componentes tienen información de las diferentes variables originales, es decir, se componen de las varianzas de las variables originales.

PC1, PC2... PCN no son variables

##Eliminación de Outliers

En la imagen que se muestra a continuación podemos observar los outliers de las variables que vamos a utilizar para nuestro modelo, sin embargo, antes de comenzar a estandarizar los datos debemos eliminar los outliers porque estos nos pueden generar sesgos importantes en los modelos de ML.

```python
Out_l =scaled_df.boxplot(figsize=(15,10), showmeans = True)

Out_l.plot()

plt.show()
```

```python
Q1=scaled_df.quantile(0.25)
print("Primer Cuartil", Q1)

Q3=scaled_df.quantile(0.75)
print("Tercer Cuartil", Q3)

IQR= Q3- Q1
print("Rango Intercuartil", IQR)
```

Calcular los valores de los bigotes inferior y superior

```python
BI_Calculado = (Q1- 1.5* IQR) #Ubicación del bigote inferior
print("BI Calculado \n", BI_Calculado)  

BS_Calculado = (Q3+ 1.5* IQR)
print("BS Calculado \n", BS_Calculado)
```

Ubicamos las filas en las que se encuentran los outliers





```python
ubc_outliers = (scaled_df< BI_Calculado) | (scaled_df>BS_Calculado)

print("\n Ubicación de outliers \n", ubc_outliers)
```

```python
outliers = scaled_df[ubc_outliers]
print("\n Lista de Outliers \n", outliers)
```

```python
outliers_ordenados = outliers.sort_values(by='Monto')

outliers_ordenados
```

Nos quedamos con los valores que no son outliers

```python
ubc_sin_outliers = (scaled_df>= BI_Calculado) & (scaled_df<= BS_Calculado)

sin_outliers = scaled_df[ubc_sin_outliers]

sin_outliers
```

```python
Out_l = sin_outliers.boxplot(figsize=(15,10), showmeans = True)
Out_l.plot()
plt.show()
```

Se usa la técnica para eliminación de outliers, algo muy similiar a lo que hace las transformación robusta para aminorar un poco su impacto, pero no siempre con buenos resultados da buenos resultados. Por ejemplo, en las variables de importe de extracto de cuenta es poco sensible a esta transformación porque su distribución tiene valores que  talvés no son "errores", en estos casos es importante conocer la opinión de los expertos en este caso. Más adelante, podemos usar transformaciones robusta para tratar de aminorar el efecto de los outliers.

5. ¿Cuándo se recomienda realizar un PCA y qué beneficios ofrece para Machine Learning?

>En machine learning es normal encontrarnos con problemas donde tengamos una enorme cantidad de features en donde hay relaciones complejas entre ellos y con la variable que queremos predecir. Se recomienda usar PCA cuando:
* Nuestro dataset tiene un número alto de features y no todos sean significativos.
* Hay una alta correlación entre los features o las relaciones entre las variables no sean linealmente separables.
* Cuando hay overfiting.
* Cuando implica un alto costo computacional.

>Se recomienda usar PCA porque nos ayuda a seleccionar los features más relevantes para el conjunto de datos, puesto si introducimos demasiados features y estos tienen valores faltantes, se harán sesgos muy significativos y vamos a perder capacidad de predicción. Hacer una buena selección de nuestro features, hará que nuestros algoritmos corran de una manera más eficiente.
