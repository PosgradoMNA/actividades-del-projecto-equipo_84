#Parte 1: Fundamentos de bases de datos

*Fundamentos de bases de datos y para ciencia de datos.  

> Podemos definir base datos como el conjunto de información o datos almacenados de forma estructurada y organizada en su mayoría guardada de forma digital, las bases de datos es una necesidad humana que se origino para almacenar la información y preservarla de las perdidas y el deterioro, para después procesar esta información y analizarla y así gestionar informes, estudios, tomar decisiones etc. Hoy en día nos relacionamos a diario con bases de datos como los son las bibliotecas, los directorios, historiales médicos, historias laborales y otras mas especializadas.

>Por otro lado, las bases de datos se dividen en relacionales y no relacionales, para la primera se usa normalización de base de datos y para la segunda se usa otro tipo de distribución de la información. Por ejemplo, SQL usa las bases de datos relacionadas y NoSQL usa la base de datos no relacionados.

>En conclusión, para la ciencia de datos el uso de las bases de datos es primordial porque el 80% del análisis se basa en la limpieza y filtado de datos a utilizar, dado que dependiendo de este análisis así será la calidad de los productos generados.

*Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.
>Tanto las bases de datos como el almacen de datos hacen referencia al almacenamiento de datos o se usan para guardar información para su posterior procesamiento y análisis. Sin embargo, los almacenes de datos se diferencian porque son capaces de almacenar información en grandes volúmenes y de diferentes fuentes en cambio la base de datos lo hacen de manera específica. En otras palabras, el almacen de datos es un conjunto de datos en grandes volumnes que provienen de distintas fuentes de información y que son centralizadas en un unico sitio para poder tener acceso desde distintas herramientas.

>Por ejemplo, una empresa puede almacenar los datos del departamento de mantenimiento en una base datos, pero para almacenar los datos de todos los departamentos debe usar un almacén de datos.

>Por otro lado, es importante aclarar que el almacen de datos no limpia la información que esta ingresando, para esto se deben realizar las tecnicas de limpieza.

####De acuerdo a las estadisticas evaluadas promedio, moda y mediana evaluadas y el comportamiento de los datos de la columa 'Edu' procedemos a imputar la adecuada con el objetivo de dejar los datos de la columna 'Edu' entre 1 y 4 y la columna 'Est_civil' reemplazar los valores de 0.

```python
new3_df=new2_df.copy()
new3_df #hago copia de la copia para realizar imputar con estadistica
```

```python
#Se puede reemplazar valores de la columna con una condición
# new_df['Edu']=new_df['Edu'].apply(lambda x:2 if x>4 else x)
# new_df['Edu']=new_df['Edu'].apply(lambda x:1 if x<1 else x)

#se puede hacer de esta forma:
#Se reemplazan los valores mayores a 4 por la mediana 2
for x in new3_df.index:
  if new3_df.loc[x, 'Edu'] > 4:
     new3_df.loc[x, 'Edu'] = 2

#Se reemplazan los valores menores a 1 por el promedio porque esta cerca a 0
for x in new3_df.index:
  if new3_df.loc[x, 'Edu'] < 1:
     new3_df.loc[x, 'Edu'] = 1
```
